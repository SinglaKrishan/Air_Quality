```python
x=b.drop(b.columns[1],axis=1)
x
```

```python
y=b.drop(b.columns[0],axis=1)
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
import math
import statistics
import operator
```

```python
X = x_train.values
X
```

```python
mean_x = np.mean(X)
mean_x
```

```python
Y=y_train.values
Y
```

```python
mean_y = np.mean(Y)
mean_y
```

```python
m=len(X)
m
```

```python
numer=0
denom=0
for i in range(m):
  numer+=(X[i] - mean_x)*(Y[i] - mean_y)
  denom+=(X[i] - mean_x) ** 2
slope = numer / denom
slope
```

```python
intercept = mean_y -(slope*mean_x)
intercept
```

```python
X_test = x_test.values
X_test
```

```python
Y_pred=slope*X_test +intercept
Y_pred
```

```python
x_test['Predicted RH']=Y_pred
x_test
```

```python
x_test['Actual RH']=y_test
x_test
```

```python
from sklearn.metrics import r2_score
score =r2_score(y_test,Y_pred)
print("The accuracy of our model is {}%".format(round(score,2)*100))
```
